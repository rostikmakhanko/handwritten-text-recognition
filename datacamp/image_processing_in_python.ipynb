```
from skimage import data
from skimage import color
from matplotlib import pyplot as plt

def show_image(image, title, cmap_type='gray'):
  plt.imshow(image, cmap=cmap_type)
  plt.title(title)
  plt.axis('off')
  plt.show()

rocket = data.rocket()

# Convert the image to grayscale
gray_scaled_rocket = color.rgb2grey(rocket)

# Show the original image
show_image(rocket, 'Original RGB image')

# Show the grayscale image
show_image(gray_scaled_rocket, 'Grayscale image')

type(rocket)
```

```
red = rocket[:, :, 2]
plt.hist(red.ravel(), bins = 256)
plt.title('Blue Histogram')
plt.show()
```

```
import numpy as np

flipped_rocket = np.flipud(rocket)
show_image(flipped_rocket, 'Flipped rocket')
```

```
# Import the otsu threshold function
from skimage.filters import threshold_otsu
# Obtain the optimal threshold value with otsu
thresh = threshold_otsu(gray_scaled_rocket)

# Apply thresholding to the image
binary = gray_scaled_rocket > thresh

# Show the image
show_image(binary, 'Binary image')
show_image(gray_scaled_rocket, 'Gray')
```

```
# Import the local threshold function
from skimage.filters import threshold_otsu, threshold_local

# Set the block size to 35
block_size = 35

# Obtain the optimal local thresholding
local_thresh = threshold_local(gray_scaled_rocket, block_size, offset=10)

# Obtain the binary image by applying local thresholding
binary_local = gray_scaled_rocket > local_thresh

# Show the binary image
show_image(gray_scaled_rocket, 'Local thresholding')
```

```
# Import the try all function
from skimage.filters import try_all_threshold

# Import the rgb to gray convertor function 
from skimage.color import rgb2gray

# Turn the fruits image to grayscale
grayscale = rgb2gray(rocket)

# Use the try all method on the grayscale image
fig, ax = try_all_threshold(grayscale, verbose=False)

# Show the resulting plots
plt.show()
```
